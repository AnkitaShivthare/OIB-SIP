```python
# unemployment rate according to different regions of India
data.columns=["States","Dates","Frequency",
              "Estimated Unemployment Rate (%)","Estimated Employed",
             "Estimated Labour Participation Rate (%)",
            "Region","longitude","latitude"]
plt.figure(figsize=(8,6))
plt.title("Unemployment in India")
sns.histplot(x="Estimated Unemployment Rate (%)", hue="Region", data=data)
plt.show()
```

```python
freq=data["Region"].value_counts()
freq
```

```python
freq.plot(kind='pie', startangle=90)
plt.legend()
plt.show()
```

### From the above pie diagram the north region is highly considered region

```python
sns.barplot(y=data["Estimated Unemployment Rate (%)"],x=data["Region"])
```

### From the above bar diagram it show that north region has the highest unemployment rate.
